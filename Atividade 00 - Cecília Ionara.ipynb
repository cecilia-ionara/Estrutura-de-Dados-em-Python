### Questão 1
Escreva um código que recebe um número inteiro e verifica se ele é par ou ímpar.

```python
num = int(input("Digite o número que deseja verificar se é par ou impar: "))

if (num%2==0):
    print(f"O número {num} é par")
else:
    print(f"O número {num} é ímpar")
```

### Questão 2
Escreva um código em Python que receba **2** valores numéricos inteiros digitados pelo usuário. O código deverá verificar qual dos **dois** números é o maior, e então mostrá-lo na tela para o usuário.


```python
num1 = int(input("Digite o 1° número: "))
num2 = int(input("Digite o 2° número: "))

if (num1>num2):
    print(f"{num1} é maior que {num2}")
elif (num2>num1):
    print(f"{num2} é maior que o {num1}")
else:
    print(f"Os números {num1} e {num2} são iguais")
```

### Questão 3
Escreva um código em Python que receba **3** valores numéricos inteiros digitados pelo usuário. O código deverá verificar qual dos **três** números é o maior, e então mostrá-lo na tela para o usuário.


```python
num1 = int(input("Digite o 1° número: "))
num2 = int(input("Digite o 2° número: "))
num3 = int(input("Digite o 3° número: "))

#usando a função max, já vem importada em python
maior = max(num1,num2,num3)

print(f"O maior valor entre os três é: {maior}")
```

### Questão 4
Escreva um código em Python que receba **4** valores numéricos inteiros digitados pelo usuário. O código deverá verificar qual dos **quatro** números é o maior, e então mostrá-lo na tela para o usuário.

```python
num1 = int(input("Digite o 1° número: "))
num2 = int(input("Digite o 2° número: "))
num3 = int(input("Digite o 3° número: "))
num4 = int(input("Digite o 2° número: "))

#usando a função max, já vem importada em python
maior = max(num1,num2,num3, num4)

print(f"O maior valor entre os quatro é: {maior}")
```

### Questão 5
Escreva um código em Python que receba **n** valores numéricos inteiros digitados pelo usuário. O código deverá verificar qual dos **n** números é o maior, e então mostrá-lo na tela para o usuário.

```python
maior = 0

quant = int(input("Digite a quantidade de números desejada para a verificação: "))

for i in range(quant):
    num = int(input(f"{i+1}° Número: "))
    if num > maior:
        maior = num

print(f"O maior valor é: {maior}")
```

# Nova seção
Uma empresa de tecnologia possui 8 programadores em início de carreira em seu quadro de profissionais. O salário de cada um deles é de R$ 4500.

A partir de dezembro, todos esses programadores passarão a receber mensalmente um aumento de 15% no salário, que é o reajuste anual, e ainda terão direito ao 13º salário, referente a dezembro, no valor de R$ 4500.


Escreva um programa que mostre:

1. Quanto cada programador irá receber no mês de dezembro, incluindo o 13º salário e o reajuste de 15%.
2. Como ficará o valor do salário mensal de cada programador nos meses seguintes.


```python
#salário é de R$ 4500.
#mensalmente um aumento de 15% no salário
#ainda terão direito ao 13º salário, referente a dezembro, no valor de R$ 4500.

salario = 4500
reajuste = 0.15 * 4500

salario_reajuste = salario + reajuste
salario_dezembro = salario_reajuste + salario #salario vale o mesmo q 13°

print(f"Em dezembro, o programador deverá receber ao todo {salario_dezembro}")

#salario sem reajuste
salario_mensal = 4500
for i in range(12):
    #conforme o laço repete o salario aumenta
    reajuste = salario_mensal * 0.15
    salario_mensal = salario_mensal + reajuste
    print(f"Mês {i+1} receberá um salário de R$ {salario_mensal:.2f}")
```
