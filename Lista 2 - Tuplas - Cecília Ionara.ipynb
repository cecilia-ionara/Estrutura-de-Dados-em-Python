Questão 1: Pratos do restaurante
Elabore um programa em Python com o seguinte contexto. Um restaurante do tipo
buffet livre oferece cinco tipos de pratos todos os dias. Na segunda-feira, por exemplo,
são oferecidos frango, feijão, quiabo, arroz e pimentão.
Crie um cardápio para ser oferecido de terça-feira a sexta-feira. Para cada dia, pense em
cinco pratos simples e armazene-os em uma tupla.
No seu programa, use um laço for para exibir cada prato oferecido pelo restaurante ao
longo da semana. O seu programa deve permitir que o cliente selecione o prato
desejado.
A escolha do cliente deve ser mostrada na tela.
Elabore as seguintes funções para o seu programa:
a) tem_prato(‘string’): a função recebe o nome de uma comida e verifica se ela conta
na tupla com os scinco pratos. A função retorna falso ou verdadeiro como resultado.
b) quantidade_pratos(): a função verifica a quantidade de pratos/comidas que constam
na tupla e retorna um número inteiro referente a quantidade encontrada.

```python
#Questao 1

def tem_prato(nome_prato, cardapio):
    if nome_prato in cardapio:
        return True
    else:
        return False

def quantidade_pratos(cardapio):
    return len(cardapio) #len = tamanho

segunda = ('frango', 'feijão', 'quiabo', 'arroz', 'pimentão')
terca = ('bife', 'batata', 'salada', 'arroz', 'tomate')
quarta = ('frango', 'feijão', 'abobrinha', 'arroz', 'alface')
quinta = ('peixe', 'purê', 'salada', 'arroz', 'cenoura')
sexta = ('carne', 'feijão', 'abóbora', 'arroz', 'beterraba')

dias_da_semana = [segunda, terca, quarta, quinta, sexta]
dias = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta']

for i in range(5):
    print(f'Cardápio da {dias[i]}: {dias_da_semana[i]}')

escolha = input("\nEscolha o prato desejado: ").lower() #lower deixa minusculas

encontrado = False #inicia zerado

for dia in dias_da_semana:
    if tem_prato(escolha, dia):
        encontrado = True
        print(f'O prato "{escolha}" está disponível!')
        break

if not encontrado:
    print(f'O prato "{escolha}" não está disponível no cardápio da semana.')

for i in range(5):
    print(f'{dias[i]}: {quantidade_pratos(dias_da_semana[i])}')
```

Questão 2: Tuplas de valores do usuário
Escreva um programa que receba como entrada do usuário uma sequência de 4 tuplas,
cada uma formada por 3 valores numéricos. O seu programa deve formar uma lista com
todas essas tuplas e mostrar na tela as seguintes informações:
a) Qual a posição na lista da tupla que apresenta a maior soma de valores?

b) Substitua as últimas posições das tuplas pela maior soma que você encontrou no
item (a).

```python
#Questao 2

tuplas = []

for i in range(4):
    tupla = tuple(map(int, input(f"Digite os 3 valores da tupla {i+1} separados por espaço: ").split())) #map cria um iterador
    tuplas.append(tupla)

somas = [sum(tupla) for tupla in tuplas]

indice_maior_soma = somas.index(max(somas))

print(f'A tupla com a maior soma está na posição {indice_maior_soma + 1}.')

#adicionar a b)
```

Questão 3: Sem tuplas vazias
Escreva uma função em Python para remover tuplas vazias de uma lista de tuplas. O seu
programa deverá conter funções que respondam o seguinte:
a) Informe em qual/quais posições da lista existem tuplas unitárias;
b) Elimine todas as tuplas vazias em uma lista de tuplas;
No seu programa, produza um teste que mostre como funciona a sua função. Para isso
utilize os seguintes dados:
Exemplo de Entrada Exemplo de Saída
[(), (), ('',), ('a', 'b'), ('a', 'b', 'c'), ('d')]

[ ('',), ('a', 'b'), ('a', 'b', 'c'), 'd']

```python
# Questão 3
def posicoes_unitarias(lista):
    posicoes = []
    for i, tupla in enumerate(lista):
        if len(tupla) == 1:  # Tupla unitária
            posicoes.append(i)
    return posicoes

def remover_tuplas_vazias(lista):
    return [tupla for tupla in lista if tupla != ()]

entrada = [(), (), ('',), ('a', 'b'), ('a', 'b', 'c'), ('d')]

posicoes = posicoes_unitarias(entrada)
print(f"Posições das tuplas unitárias: {posicoes}")

saida = remover_tuplas_vazias(entrada)
print(f"Lista após remover as tuplas vazias: {saida}")
```

Questão 4: Dados do usuário
Elabore um programa que receba como entrada cinco conjuntos de dados de usuários,
os quais serão representados por cinco tuplas contendo nome, idade e cidade de
nascimento, cujos dados são fornecidos pelo usuário. Todas as tuplas devem ser
organizadas como elementos em uma lista. Em seguida responda:
a) Ordene a lista pelos nomes nas tuplas e informe todos os nomes existentes;
b) Ordene a lista pelas idades e mostre todas as idades;
c) Ordene a lista pela cidade de nascimento e mostre todas as cidades.*Texto em itálico*

```python
#Questao 4
dados_usuarios = []

for i in range(5):
    nome = input(f"Digite o nome do usuário {i+1}: ")
    idade = int(input(f"Digite a idade do usuário {i+1}: "))
    cidade = input(f"Digite a cidade de nascimento do usuário {i+1}: ")

    dados_usuarios.append((nome, idade, cidade))

dados_usuarios.sort(key=lambda x: x[0]) #key lambda: função para extrair um valor da tupla // indice
print("Nomes ordenados:")
for dado in dados_usuarios:
    print(dado[0])

dados_usuarios.sort(key=lambda x: x[1])
print("\nIdades ordenadas:")
for dado in dados_usuarios:
    print(dado[1])

dados_usuarios.sort(key=lambda x: x[2])
print("\nCidades ordenadas:")
for dado in dados_usuarios:
    print(dado[2])
```

Na linguagem Python, os tipos de dados são usados para classificar um tipo específico
de dados, determinando os valores que você pode atribuir ao tipo e as operações que
você pode realizar nele. Às vezes, dependendo do problema computacional que se quer
resolver, é necessário converter um tipo de estrutura em outra. Para facilitar a vida dos
programadores, uma equipe de alunos planeja disponibilizar na internet uma ferramenta
que faça a conversão de entradas de dados para a estrutura de dados tupla. O objetivo da
equipe é que a saída a ser impressa na tela para os usuários seja sempre uma tupla. Para
isso, a equipe de alunos deseja que você elabore um programa que receba, em uma
única linha, uma sequência de números inteiros digitada pelo usuário, e que manipule
essa sequência tornando-a uma tupla, que deve ser impressa na tela ao final. Mas existe
uma restrição: a tupla que será impressa não poderá conter números repetidos. Por isso,
a tupla resultante poderá ter os mesmos números digitados na entrada, ou até menos
números.

```python
#Questao 5
entrada = input()

numeros = [int(x) for x in entrada.split()] #split remove os espaços

tupla = tuple(set(numeros)) #set remove os numeros repetidos #tuple é pra voltar a ser tupla

print(tupla)
```
